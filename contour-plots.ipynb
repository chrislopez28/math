```python
from mpl_toolkits import mplot3d
```

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')
```

```python
def f(L, K):
    return 1.01 * (L ** 0.75) * (K ** 0.25)
```

```python
l = np.linspace(0, 300)
k = np.linspace(0, 300)
```

```python
L, K = np.meshgrid(l, k)
```

```python
P = f(L, K)
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')

ax.contour3D(L, K, P, 50, cmap='binary')
ax.set_xlabel('L')
ax.set_ylabel('K')
ax.set_zlabel('P');
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')

ax.contour3D(L, K, P, 50, cmap='binary')

ax.invert_xaxis()
ax.set_xlabel('L')
ax.set_ylabel('K')
ax.set_zlabel('P');
```
