```python
import numpy as np
import matplotlib.pyplot as plt
```

Assuming 1st-order growth, we can model the change of Population over time using the following differential equation:

$$ \frac{dP}{dt} = kP $$

Solving this differential equation for P gives:

$$ P = e^{\ln{P_0} + kt}$$

If P is known at two different times, we can solve for k:

$$ k = \frac{\ln{P} - \ln{P_0}}{t} $$

```python
def P(t):
    return np.exp(np.log(4E9) + (0.006931472*t))
```

```python
P(100)
```

```python
t = np.arange(0.0, 150, 1) 
plt.plot(t, P(t))
plt.xlabel('Years after Today')
plt.ylabel('Population')
plt.title('Modeled Population Growth over next 150 Years')
```
